```python
sentence="Sundar Pichai, the CEO of Google, announced a new AI initiative in California on Monday."

"""
Possible entities:

Sundar Pichai → PERSON
Google → ORGANIZATION
California → LOCATION/GPE
Monday → DATE"""
```

```python
sentence="Sundar Pichai, the CEO of Google, announced a new AI initiative in California on Monday."
```

```python
import nltk
nltk.download('punkt_tab')
words=nltk.word_tokenize(sentence)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
tag_elements=nltk.pos_tag(words)
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.download('words')
```

```python


named_entities = nltk.ne_chunk(tag_elements)
display(named_entities)
```
